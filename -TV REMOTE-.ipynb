```python
import random
import time


class remote():

    def __init__(self, tv="open", tv_voice=0, channel_list=["BBC"], channel="BBC", connection="connected"):
        self.tv = tv
        self.tv_voice = tv_voice
        self.channel_list = channel_list
        self.channel = channel
        self.connection = connection

    def open_tv(self):

        if (self.tv == "open"):
            print("TV is already on")
        else:
            print("TV turns on")
            self.tv = "open"

    def close_tv(self):

        if (self.tv == "open"):
            print("TV turns off")
            self.tv = "close"
        else:
            print("TV is already off")

    def open_int(self):

        if (self.connection == "connected"):
            print("Internet connected")
        else:
            print("Internet disconnects")

    def close_int(self):

        if (self.connection == "connected"):
            print("Internet disconnects")
        else:
            print("Internet not already connected")

    def voice_sett(self):

        while True:
            answer = input("turn down the volume: '<'\nturn up the volume: '>'\nexit: exit :")

            if (answer == "<"):
                if (self.tv_voice != 0):
                    self.tv_voice -= 1
                    print("voice:", self.tv_voice)
            elif (answer == ">"):
                if (self.tv_voice != 50):
                    self.tv_voice += 1
                    print("voice:", self.tv_voice)
            else:
                print("volume updated", self.tv_voice)
                break

    def add_channel(self, channel_name):
        print("Adding new channel to tv list")
        time.sleep(1)

        self.channel_list.append(channel_name)
        print("Added channel named", channel_name)

    def random_channel(self):
        randoms = random.randint(0, len(self.channel_list) - 1)
        self.channel = self.channel_list[randoms]
        print("Channel is", self.channel)

    def __len__(self):
        return len(self.channel_list)

    def __str__(self):
        return "tv = {}\nvoice = {}\nchannel_list = {}\nChannel is = {}\n".format(self.tv,self.tv_voice,self.channel_list,self.channel)

remote= remote()

print("""

SMART TELEVISION APP
1- Open TV
2- Close TV
3- Connect to Internet
4- Disconnect Internet
5- TV Voice settings
6- Add chanel
7- Number Of Channels
8- Random Channel
9- Information Of The Television

PRESS 'q' TO EXIT

 """)

while True:

    operation = input("Select The operation:")

    if (operation == "q"):
        print("Smart TV app is being terminated")
        break

    elif (operation == "1"):
        remote.open_tv()

    elif (operation == "2"):
        remote.close_tv()

    elif (operation == "3"):
        remote.open_int()

    elif (operation == "4"):
        remote.close_int()

    elif (operation == "5"):
        remote.voice_sett()

    elif (operation == "6"):
        channel_names = input("Type the channel names by typing the ' , ' sign between the names : ")

        channel_list = channel_names.split(",")

        for additive in channel_list:
            remote.add_channel(additive)

    elif (operation == "7"):
        print("Number of channels", len(remote))

    elif (operation == "8"):
        remote.random_channel()

    elif (operation == "9"):
        print(remote)

    else:
        print("Invalid input")
        








```
